```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
%matplotlib inline

df = pd.read_csv('./survey_results_public.csv')
df.head()
```

```python
columns_of_interest = [
    'Professional',
    'Country',
    'University',
    'EmploymentStatus',
    'DeveloperType',
    'HaveWorkedLanguage',
    'WantWorkLanguage',
    'Gender',
    'Salary'
]

sub_df = df.loc[:, columns_of_interest] 
```
